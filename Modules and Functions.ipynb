```python
import time
from time import time as my_timer
import random
input("press enter to start")
wait_time = random.randint(12, 60)
time.sleep(wait_time)
start_time = my_timer()
end_time = my_timer()
print("started"+ time.strftime("%X", time.localtime(start_time)))
print("stop time"+ time.strftime("%X", time.localtime(end_time)))
```
